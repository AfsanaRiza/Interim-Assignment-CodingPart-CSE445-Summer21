# Import Libraries

```python
import numpy as np
import cv2
import tensorflow as tf

import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
input_dir = '/content/drive/MyDrive/Segmentation/test'
```

```python
datagen = ImageDataGenerator()
test_it = datagen.flow_from_directory('/content/drive/MyDrive/Segmentation', class_mode='binary', batch_size=40)
```

```python
original_image_1 = cv2.imread("/content/drive/MyDrive/Segmentation/test/12003.jpg")
original_image_2 = cv2.imread("/content/drive/MyDrive/Segmentation/test/25098.jpg")
```

## Segmentation For Image_1

```python
img=cv2.cvtColor(original_image_1,cv2.COLOR_BGR2RGB)

vectorized = img.reshape((-1,3))
vectorized = np.float32(vectorized)

K_1 = 3
attempts=10
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret,label,center=cv2.kmeans(vectorized,K_1,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_original_image_11 = res.reshape((img.shape))

K_2 = 5
attempts=10
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret,label,center=cv2.kmeans(vectorized,K_2,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_original_image_12= res.reshape((img.shape))

K_3 = 7
attempts=10
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret,label,center=cv2.kmeans(vectorized,K_3,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_original_image_13 = res.reshape((img.shape))


figure_size = 20
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(1,4,1),plt.imshow(img)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(1,4,2),plt.imshow(result_original_image_11)
plt.title('Segmented Image when K = %i' % K_1), plt.xticks([]), plt.yticks([])
plt.subplot(1,4,3),plt.imshow(result_original_image_12)
plt.title('Segmented Image when K = %i' % K_2), plt.xticks([]), plt.yticks([])
plt.subplot(1,4,4),plt.imshow(result_original_image_13)
plt.title('Segmented Image when K = %i' % K_3), plt.xticks([]), plt.yticks([])
plt.show()
```
